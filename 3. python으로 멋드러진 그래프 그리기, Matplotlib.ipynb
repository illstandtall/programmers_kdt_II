#### numpy.array를 이용해서 함수 그래프 그리기

- $y=x^2$

```python
x = np.array([1, 2, 3, 4, 5]) # domain
y = np.array([1, 4, 9, 16, 25]) # range

plt.plot(x, y)
plt.show()
```

##### np.arange(a, b, c)

- c: 0.01 만큼 증가하며 값을 채운다.

```python
x = np.arange(-10, 10, 0.01)
x[:5]
```

```python
plt.plot(x, x**2)
plt.show()
```

##### x, y축에 설명 추가하기

- `plt.xlabel`: x 축의 이름을 정할 수 있다.
- `plt.label`: y 축의 이름을 정할 수 있다.

```python
x = np.arange(-10, 10, 0.01)

plt.xlabel('x value')
plt.ylabel('f(x) value')

plt.plot(x, x**2)
plt.show()
```

##### x, y축의 범위를 설정하기

- `plt.axis`: `[x_min, x_max, y_min, y_max]`로 범위를 정할 수 있음

```python
x = np.arange(-10, 10, 0.01)

plt.xlabel('x value')
plt.ylabel('f(x) value')

plt.axis([-5, 5, 0, 25])

plt.plot(x, x**2)
plt.show()
```

##### x, y축의 눈금 설정하기

- `plt.xticks`: `x`축 눈금 설정
- `plt.yticks`: `y`축 눈금 설정

```python
x = np.arange(-10, 10, 0.01)


plt.xlabel('x value')
plt.ylabel('f(x) value')

plt.axis([-5, 5, 0, 25])

plt.xticks([i for i in range(-5, 5, 1)])
plt.yticks([i for i in range(0, 27, 3)])

plt.plot(x, x**2)
plt.show()
```

##### 그래프에 title 갈기

- `plt.title`: 그래프에 제목을 설정할 수 있다

```python
x = np.arange(-10, 10, 0.01)


plt.xlabel('x value')
plt.ylabel('f(x) value')

plt.axis([-5, 5, 0, 25])

plt.xticks([i for i in range(-5, 5, 1)])
plt.yticks([i for i in range(0, 27, 3)])

plt.title("y=x^2 graph")

plt.plot(x, x**2)
plt.show()
```

##### 그래프 선에 이름 달기

1. `plt.plot(label='name')`: 이름을 달아 준 후
2. `plt.legend()`:를 사용한다.

```python
x = np.arange(-10, 10, 0.01)


plt.xlabel('x value')
plt.ylabel('f(x) value')

plt.axis([-5, 5, 0, 25])

plt.xticks([i for i in range(-5, 5, 1)])
plt.yticks([i for i in range(0, 27, 3)])

plt.title("y=x^2 graph")

plt.plot(x, x**2, label="Trend")

plt.legend()
plt.show()
```

## III. Matplotlib Case Study

### 꺾은 선 그래프 (Plot)

- 시계열 데이터를 나타내는 데에 많이 사용
- 데이터들의 분포
- `plt.plot` 사용

- numpy.random.randint(a, b, ...)는 [a, b) 의 범위 중 하나의 수를 임의로 추출합니다.
- cf) random.randint(a, b)는 [a, b]의 범위 중 하나의 수를 임의로 추출합니다.

```python
x = np.arange(20) # 0~19
y = np.random.randint(0, 20, 20) # 난수를 20번 생성

x, y
```

```python
plt.plot(x, y)
plt.show()
```

#### extra mission

- 'y'축을 20까지 보이게 하고 싶다면? `plt.axis`
- 'y'축을 5단위로 보이게 하고 싶다면? `plt.yticks`

```python
plt.yticks([0, 5, 10, 15, 20])
plt.axis([0, 20, 0, 20])

plt.plot(x, y)
plt.show()
```

### 산점도 (Scatter Plot)

- 선으로 잇지 않고 점으로만 표현할 수 있다.
- 서로 떨어진 점 들이 어떠한 그래프를 만드는지(양/음의 상관관계) 파악에 좋다.
- 데이터들의 분포
- `plt.scatter()`를 사용


```python
plt.scatter(x, y)
plt.show()
```

### 박스 그림 (Box Plot)

- 수치형 데이터에 대한 정보 (Q1, Q2, Q3, min, max) 
- 데이터의 축약된 정보
- `plt.boxplot()`

```python
plt.boxplot(y)

plt.show()
```

#### extra mission

- box의 title 적용하기

```python
plt.boxplot(y)
plt.title('Box plot of y')

plt.show()
```

```python
plt.boxplot((x, y))
plt.title('Box plot of x, y')

plt.show()
```

### 막대 그래프 (Bar Plot)

- 범주형 데이터의 '값'과 그 '값의 크기'를 직사각형으로 나타낸 그림
- 데이터들의 도수
- ex) 남성/여성, 점수분포
- `plt.bar()`

```python
plt.bar(x, y)

plt.show()
```

#### extra mission

- `xticks`를 올바르게 처리하기

```python
plt.bar(x, y)

plt.xticks(np.arange(0, 20))

plt.show()
```

#### cf) Histogram

- 도수분포를 직사각형의 막대 형태로 나타낸 것
- '계급'으로 나타낸 것이 특징
- 0, 1, 2가 아니라 0~2까지의 '범주형'데이터로 구성 후 그림을 그린다.
- 범주가 이어져 있기 때문에 그래프가 이어져 있음
- 데이터들의 도수
- `plt.hist` 사용

```python
plt.hist(y, bins=np.arange(0, 22, 2))

plt.show()
```

##### extra mission

- `xticks`를 올바르게 처리하기

```python
plt.hist(y, bins=np.arange(0, 22, 2))

plt.xticks(np.arange(0, 22, 2))

plt.show()
```

### 원형 그래프 (Pie Chart)

- 데이터에서 전체에 대한 부분의 비율을 부채꼴로 나타낸 그래프
- 다른 그래프에 비해서 **비율** 확인에 용이
- 비율!
- `plt.pie()` 사용

```python
z = [100, 300, 200, 400]

plt.pie(z)
plt.show()
```

```python
z = [100, 300, 200, 400]

plt.pie(z, labels=['100', '200', '300', '400'])
plt.show()
```

### Seaborn Import하기

```python
import seaborn as sns
```

### 커널 밀도그림 (Kernel Density Plot)

- 히스토그램과 같은 연속적인 분포를 곡선화해서 그린 그림
- `sns.kdeplot()`

#### in Histogram


```python
x = np.arange(0, 22, 2)
y = np.random.randint(0, 20, 20)

plt.hist(y, bins=x)
```

#### in kdeplot

```python
sns.kdeplot(y, shade=True)

plt.show()
```

### 카운트 그림 (Count Plot)

- 범주형 Column(category)의 빈도수를 시각화
- Groupby 후의 도수를 하는 것과 동일한 효관
- `sns.countplot()` 사용

```python
vote_df = pd.DataFrame({"name": ['Andy', 'Bob', 'Cat'], "vote": [True, True, False]})

vote_df
```

#### in barplot

```python
vote_count = vote_df.groupby('vote').count()

vote_count
```

```python
plt.bar(x=[False, True], height=vote_count['name'])

plt.show()
```

#### in sns.countplot

```python
sns.countplot(x=vote_df['vote'])

plt.show()
```

### 캣 그림(Cat Plot)

- 숫자형 변수와 하나 이상의 범주형 변수의 관계를 보여주는 함수
- `sns.catplot()` 사용
- 복잡한 데이터에 유리함
- 범주형 데이터(x)와 수치형 데이터(y)를 매핑하는데 사용하는 방법으로 보는 것이 적절
- 기본적으로 **strip plot**으로 그려줌

```python
covid = pd.read_csv("./country_wise_latest.csv")

covid.head(5)
```

```python
sns.catplot(x='WHO Region', y='Confirmed', data=covid)

plt.show()
```

```python
s = sns.catplot(x='WHO Region', y='Confirmed', data=covid) # kind=defalut: strip
s.fig.set_size_inches(10, 6)

plt.show()
```

### 스트립그림 (Strip Plot)

- scatter plot과 유사하게 데이터의 수치를 표현하는 그래프
- `sns.stripplot()` 사용

```python
sns.stripplot(x='WHO Region', y='Recovered',  data=covid)

plt.show()
```

#### cf) swarpplot

- 점들이 겹치는 경우 양 옆으로 분산합니다.
- `sns.swarmplot()`
- 밑에 예의 경우 에러가 나는 것은 데이터가 많아 모두 표현하기 어렵다는 것이므로 무시해도 됩니다.

```python
sns.swarmplot(x='WHO Region', y='Recovered',  data=covid)

plt.show()
```

### 히트맵 (Heatmap)

- 데이터의 행렬을 색상으로 표현해주는 그래프
- 상관계수 행렬에 많이 사용
- `sns.heatmap()` 사용


```python
covid.corr()
```

```python
sns.heatmap(covid.corr())

plt.show()
```

- 밝은 색으로 나타낸 곳이 양의 상관관계가 있음을 나타냅니다.
